```python
import altair as alt
import numpy as np
```

```python
# import functions from modules
from generate_data import generate_data
from generate_predictions import nearest_neighbor_predictor
from plot_results import scatter_plot
```

```python
num_samples = 50

rng = np.random.default_rng(seed=42)

training_data = generate_data(rng, num_samples)
test_data = generate_data(rng, num_samples)

training_data
```

```python
num_neighbors = 7

test_data["prediction"] = nearest_neighbor_predictor(
    training_data, test_data, num_neighbors
)

test_data
```

```python
# create a new column "correct" where column label matches column prediction
test_data["correct"] = test_data["label"] == test_data["prediction"]
```

```python
# number of rows where correct is True
num_correct = test_data["correct"].sum()
accuracy = num_correct / len(test_data)
print(f"Accuracy: {accuracy}")
```

```python
chart1 = scatter_plot(
    test_data, f"Are predictions correct? (accuracy: {accuracy})", "correct", "set1"
)

chart2 = scatter_plot(training_data, "Training data", "label", "set2")

# the resolve part makes sure that the legends don't get merged and that
# the color scales remain independent
combined_chart = alt.vconcat(chart1, chart2).properties(
    resolve=alt.Resolve(
        scale=alt.LegendResolveMap(color=alt.ResolveMode("independent"))
    ),
)

combined_chart
```
